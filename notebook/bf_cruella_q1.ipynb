What is the rate of re-referral? 
How many children have multiple referrals (by year or by Point of Entry Office (POE))?

```python
ref = teis_concat.groupby(['poe', 'fiscal_year'])['child_id'].count()
```

```python
xy = teis_month[teis_month.duplicated(subset = "child_id", keep = False)]
reref = xy.groupby(["poe", "fiscal_year"])["child_id"].nunique()
together = (pd.merge(ref, reref, left_on=['poe', 'fiscal_year'], right_on=['poe', 'fiscal_year'], how='left')
            .rename(columns = {'child_id_x' : 'total_referrals', 'child_id_y' : 'total_re-referrals'}))
together['percent_referrals'] = (together['total_re-referrals']/together['total_referrals']) * 100
together
```

```python
## or you can use the fiscal year.1 for a single year
teis_concat.groupby(['poe', 'fiscal_year.1'])['child_id'].count()
```

```python
ref = teis_month.groupby(["poe"])["child_id"].count()
ref1 = teis_ref.groupby(['poe'])["child_id"].count()
xy = teis_month[teis_month.duplicated(subset = "child_id", keep = False)]
xy1 = teis_ref[teis_ref.duplicated(subset = "child_id", keep = False)]
reref = xy.groupby(["poe"])["child_id"].nunique()
reref1 = xy1.groupby(["poe"])["child_id"].nunique()
together = (pd.merge(ref, reref, left_on=['poe'], right_on=['poe'], how='left')
            .rename(columns = {'child_id_x' : 'Total Referrals', 'child_id_y' : 'Total Re-referrals'}))
together1 = (pd.merge(ref1, reref1, left_on=['poe'], right_on=['poe'], how='left')
            .rename(columns = {'child_id_x' : 'Total Referrals', 'child_id_y' : 'Total Re-referrals'}))
together['Percent Re-referral'] = (together['Total Re-referrals']/together['Total Referrals']) * 100
together1['Percent Re-referral'] = (together1['Total Re-referrals']/together1['Total Referrals']) * 100
together.reset_index()
```

```python
## make a graph
```

```python
sns.set_style('white')
m = sns.FacetGrid(together, col='poe')
m.map_dataframe(sns.scatterplot, x='', y='Total Referrals')
m.set_titles(col_template='{col_name}')
```
